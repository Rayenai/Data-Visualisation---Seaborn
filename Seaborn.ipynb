```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Pre_processed_Placement.csv")
```

```python
dataset
```

```python
print("The number of rows:",dataset.shape[0])
print("The number of columns:",dataset.shape[1])
```

```python
dataset.shape
```

```python
#Univariate Analysis
# Histogram
import seaborn as sb
sb.distplot(dataset['ssc_p'],hist=True,kde=True,kde_kws={'color':'red'})
```

```python
#Bivariate Analysis
#jointplot
import seaborn as sb
sb.jointplot(x='ssc_p',y='salary',data=dataset)
plt.show()
```

```python
import seaborn as sb
sb.jointplot(x='ssc_p',y='salary',data=dataset,kind='hex')
plt.show()
```

```python
import seaborn as sb
sb.jointplot(x='ssc_p',y='salary',data=dataset,kind='kde')
plt.show()
```

```python
#pair plot
import seaborn as sb
sb.pairplot(dataset,hue='gender',diag_kind="kde",kind="scatter",palette="husl")
plt.show()
```

```python
# Categorical data vs Numerical data
#Stripplot
import seaborn as sb
sb.stripplot(x="ssc_b",y="ssc_p",data=dataset,jitter=True)
plt.show()
```

```python
#Swarmplot
import seaborn as sb
sb.swarmplot(x="ssc_b",y="ssc_p",data=dataset)
plt.show()
```

```python
#Distribution Observation
#Violinplot
import seaborn as sb
sb.violinplot(x='gender',y='ssc_p',data=dataset)
plt.show()
```

```python
import seaborn as sb
sb.violinplot(x='specialisation',y='salary',data=dataset,hue='gender')
plt.show()
```

```python
#Box plot
import seaborn as sb
sb.boxplot(x='ssc_b',y='ssc_p',data=dataset,hue='gender')
plt.show()
```

```python
#Bar Plot
import seaborn as sb
sb.barplot(x='ssc_b',y='ssc_p',data=dataset,hue='gender')
plt.show()
```

```python
#Count Plot
import seaborn as sb
sb.countplot(x='ssc_b',data=dataset,hue='gender')
plt.show()
```

```python
#Point plot
import seaborn as sb
sb.pointplot(x='ssc_p',y='hsc_b',data=dataset,hue='gender')
plt.show()
```

```python
#Box Plot
import seaborn as sb
sb.boxplot(data=dataset,orient='h')
plt.show()
```

```python
#Factorplot or Catplot
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('exercise')
sb.catplot(x='time',y='pulse',hue='kind',kind='violin',col='diet',data=df)
plt.show()
```

```python
# Reg plot & lm plot
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
sb.regplot(x='total_bill',y='tip',data=df)
sb.lmplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('anscombe')
sb.lmplot(x="x",y="y",data=df.query("dataset=='II'"))
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('anscombe')
sb.lmplot(x="x",y="y",data=df.query("dataset=='II'"),order=2)
plt.show()
```
